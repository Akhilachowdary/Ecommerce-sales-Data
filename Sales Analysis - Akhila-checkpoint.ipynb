```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pymysql
import pyodbc
```

```python
path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("all_data_copy.csv", index=False)
```

```python
df2['month'] = df2['month'].astype('int32')
df2['Quantity Ordered'] = df2['Quantity Ordered'].astype('int32')
df2['Price Each'] = df2['Price Each'].astype('float')
```

```python
df2.drop(['Date'], axis=1)
```

```python
df2['sales'] = df2['Quantity Ordered']*df2['Price Each']
df2.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df3 = df[['Order ID', 'Grouped']].drop_duplicates()

from itertools import combinations
from collections import Counter

count = Counter()

for row in df3['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
    
#('iPhone', 'Lightning Charging Cable') 1005
```

```python
df2['city'] = df2['Purchase Address'].apply(lambda x: (x.split(',')[1]))
df2['city'] = df2['city'].apply(lambda x: x.strip())
df2.drop(['Date'],axis=1)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
df3 = df2.copy()  
```

```python
df3['Order Date'] = pd.to_datetime(df3['Order Date'], format = '%m/%d/%y %H:%M')
```

```python
df4 = df3.drop(['Date'],axis=1)
```

```python
df4.rename(columns={
    'Order ID' : 'order_id',
    'Product' : 'product',
    'Quantity Ordered' : 'quantity_ordered',
    'Price Each' : 'price_each',
    'Order Date' : 'order_date',
    'Purchase Address' : 'purchase_address',
    'month' : 'month_of_purchase',
    'sales' : 'sale_per_id',
    'city' : 'city'
}, inplace = True)
```

```python
df4.head(2)
```

```python
df4.to_csv('clean_sales_data.csv', index=False)
```

```python
df4['hour'] = df4['order_date'].apply(lambda x: x.hour)
```

```python
df4.head()
```

```python
month_sales = list(df4.groupby(['month_of_purchase']).sum().sale_per_id)
```

```python
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
plt.figure(figsize = [12,5])

sns.barplot(x = months, y = month_sales,
            palette = 'hls',
            capsize = 0.05,             
            saturation = 8,             
            errcolor = 'gray', errwidth = 1,  
            ci = 'sd'   
            )
plt.show()
```

```python
product_sales = df4.groupby(['product']).sum().sale_per_id
```

```python
product_names = df4['product'].unique()
product_names
```

```python
plt.figure(figsize = [10,10])

sns.barplot(x = product_sales, y = product_names, 
             color = 'purple', edgecolor = 'w')
```
